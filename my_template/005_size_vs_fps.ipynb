```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# data
cpu = pd.read_csv('size_fps_i7-10700.csv')
gpu = pd.read_csv('size_fps_RTX-3090.csv')

# set figure
fig, ax1 = plt.subplots(dpi=100)

ax1.set_xlabel('Size', fontsize=20)
ax1.set_ylabel('FPS', fontsize=20)
ax1.grid(alpha=.5)

# plot
l1, = ax1.plot(cpu['size'], cpu['fps'], 'o-', label='CPU i7-10700')
l2, = ax1.plot(gpu['size'], gpu['fps'], 'o-', label='GPU RTX-3090')
plt.legend(handles=[l1,l2,], loc='best')

plt.show()
```
