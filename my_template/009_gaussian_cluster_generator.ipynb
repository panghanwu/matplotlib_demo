```python
import matplotlib.pyplot as plt
import numpy as np

NUM_CLUSTERS = 6
CENTER_RANGE = [-1000., 1000.]  # range of x and y
WIDTH_RANGE = [10., 300.]  # range of gaussian width
SIZE_RANGE = [100, 500]  # range of cluster sizes

# generate clusters
clusters = list()
for i in range(NUM_CLUSTERS):
    data = {
        'x': list(),
        'y': list()
    }
    
    size = np.random.randint(*SIZE_RANGE)
    for p in data:
        mu = np.random.uniform(*CENTER_RANGE)
        std = np.random.uniform(*WIDTH_RANGE)
        data[p] = np.random.normal(mu, std, size)
        
    clusters.append(data)
        

# plot
plt.figure(figsize=(6,6), dpi=100)
a = 1.5*CENTER_RANGE[0]
b = 1.5*CENTER_RANGE[1]
plt.xlim(a, b)
plt.ylim(a, b)

for clu in clusters:
    plt.scatter(clu['x'], clu['y'], alpha=0.7)

plt.show()
```
