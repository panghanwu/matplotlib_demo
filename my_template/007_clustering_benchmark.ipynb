```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
SIZE = 500

# dataset 1
MU = 0
STD = 5

d1 = dict()
d1['x'] = np.random.normal(MU, STD, SIZE)
d1['y'] = np.random.normal(MU, STD, SIZE)

# dataset 2
MU = 20
STD = 2

d2 = dict()
theda = np.random.uniform(0, 2*np.pi, SIZE)
d2['x'] = MU*np.sin(theda) + np.random.normal(0, STD, SIZE)
d2['y'] = MU*np.cos(theda) + np.random.normal(0, STD, SIZE)

# plot
plt.figure(figsize=(6,6), dpi=100)
plt.xlim(-30, 30)
plt.ylim(-30, 30)

plt.scatter(d1['x'], d1['y'], c='tab:blue', alpha=0.7)
plt.scatter(d2['x'], d2['y'], c='tab:red', alpha=0.7)

plt.show()
```
